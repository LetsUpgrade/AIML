```python
# CLeaning the texts for all review using for loop

import re
import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0,1000):
    review = re.sub( '[^a-zA-Z]' , ' ' , dataset['Review'][i] )
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ ps.stem(w) for w in review if not w in set(stopwords.words('english')) ]
    review = ' '.join(review)
    print(review)
    corpus.append(review)
```

```python
# Creating the Bag of Words Model

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
y = dataset.iloc[:,1].values
y
```

```python
# Splitting the dataset into train and test

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
# Fitting a Navive Bayes to the Training set

from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
```

```python
classifier.fit(X_train, y_train)
```

```python
# Prediction

y_pred = classifier.predict(X_test)
```

```python
# Making a Confusion Matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test , y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test , y_pred))
```

```python
Review = "nice service"
input_data = [Review]

input_data = cv.transform(input_data).toarray()

input_pred = classifier.predict(input_data)

if input_pred[0]==1:
    print("Review is Positive")
else:
    print("Review is Negative")
```
